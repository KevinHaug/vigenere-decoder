```python
# Calculate the Index of Coincidence for a given string
# Based on https://pages.mtu.edu/~shene/NSF-4/Tutorial/VIG/Vig-IOC.html
def calculate_ioc(string):
        letter_frequency = np.zeros(len(alphabet), dtype=int)
        occurrences = 0
        N = len(string)
        print(letter_frequency) # 
        for i in range(len(alphabet)):
                print(i, ":\t", alphabet[i], "\t", string.count(alphabet[i]))
                letter_frequency[i] = string.count(alphabet[i])
        print("New letter frequency: ", letter_frequency) # debug
        #    letter_frequency[i] = string.count(alphabet[i])
        #print("Letter frequency: ", letter_frequency) # debug


        # Calculating F_i (F_i - 1)

        

        for i in range(len(alphabet)):
                char_occurrence = letter_frequency[i]
                occurrences += char_occurrence * (char_occurrence - 1)

        print("Occurrences: ", occurrences) # debug

        result = occurrences / (N * (N - 1))
        print("IOC result: ", result)
        return result
```
